```python
print("List and its methods\n--------------------\n")
a=[1,2,4,2,5,5]
a.append(3)
print("After appending 3,list a is:",a)
a.insert(1,7)
print("After inserting 7 at 1st position ,list a is:",a)
a.pop(2)
print("After removing element at index 2,list a is:",a)
a.remove(2)
print("After removing the first occurrence of 2 in list,list a is:",a)
b=a.copy()
print("After copying list a to b,list b is:",b)
count=a.count(5)
print("The count of 5 in list is",count)
length=len(a)
print("The number of elements in a is",length)
a.sort()
print("The sorted list is:",a)
a.reverse()
print("The reversed list is:",a)
index=a.index(4)
print("The first index of 4 is",index)
c=[8,9,10]
a.extend(c)
print("The list a appended to list c is:",a)
a.clear()
print("After clearing list a is:",a)
print("\n")


print("Tuple and its methods\n---------------------\n")
a=(1,2,3,8,8)
length=a.count(8)
print("The number of times 8 appears in a is:",length)
index=a.index(3)
print("The element 3 appears first at index:",index)
print("The element at index 3 is:",a[3])
print("\n")


print("Dictionary and its methods\n--------------------------\n")
a={
    '1':'Python','2':'C','3':'Java'
    }
print("The value of key 1 is",a.get('1'))
print("The keys in the dictionary a are:",a.keys())
print("The values in dictionary a are:",a.values())
b={
    '1':'Kotlin','2':'C++'
  }
a.update(b)
print('After changing the the values of a,new a is:',a)
a.pop('2')
print('After removing the value of key 2,dictionary a is',a)
a.popitem()
print("After removal of random key value pair,a is:",a)
a.clear()
print("After clearing,dictionary a is:",a)
```
